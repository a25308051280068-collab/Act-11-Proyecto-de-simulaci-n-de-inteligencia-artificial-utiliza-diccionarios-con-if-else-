```python
# =====================================================================
# SIMULADOR DE IA: GESTOR DE PERFILES LABORALES (RECEPCIONISTAS)
# =====================================================================

# Lista global que guardará los diccionarios (nuestra "Base de Datos" de la IA)
base_datos_ia = []

# Función para mostrar el menú principal
def mostrar_menu():
    print("\n" + "="*50)
    print("      BRAIN-BOT: SIMULADOR DE IA (MENÚ)")
    print("="*50)
    print("1. Entrenar IA (Registrar nueva Recepcionista)")
    print("2. Consultar Base de Datos (Ver Tabla)")
    print("3. Salir")
    print("="*50)

# Función para registrar los datos (Input de Usuario)
def registrar_recepcionista():
    print("\n--- ENTRENANDO A LA IA: ENTRADA DE DATOS ---")

    # Forzamos a que el diccionario empiece vacío aquí mismo para evitar el NameError
    recepcionista = {}

    # Captura de datos (Key: Value)
    recepcionista["empresa"] = input("1. Empresa: ")
    recepcionista["idiomas"] = input("2. Idiomas hablados: ")
    recepcionista["sistema"] = input("3. Sistema de reservas que domina: ")
    recepcionista["extension"] = input("4. Extensión telefónica asignada: ")
    recepcionista["horario"] = input("5. Horario de entrada (ej. 08:00 AM): ")
    recepcionista["tareas"] = input("6. Tarea administrativa principal: ")

    # Guardamos el diccionario en nuestra lista
    base_datos_ia.append(recepcionista)
    print("\n¡Éxito! La IA ha procesado y almacenado el nuevo perfil.")

# Función para mostrar los datos en una tabla bien alineada
def mostrar_tabla():
    # El condicional 'if' verifica si la lista está vacía
    if len(base_datos_ia) == 0:
        print("\n[!] La IA no tiene datos aún. Por favor, selecciona la opción 1 primero.")
    else:
        print("\n" + "="*115)
        print(f"{'EMPRESA':<15} | {'IDIOMAS':<15} | {'SIST. RESERVAS':<18} | {'EXT. TEL':<10} | {'HORARIO':<12} | {'TAREA ADM.' :<25}")
        print("="*115)

        # Ciclo FOR para recorrer la lista de diccionarios
        for perfil in base_datos_ia:
            print(f"{perfil['empresa']:<15} | {perfil['idiomas']:<15} | {perfil['sistema']:<18} | {perfil['extension']:<10} | {perfil['horario']:<12} | {perfil['tareas']:<25}")

        print("="*115)

# =====================================================================
# FLUJO PRINCIPAL DEL PROGRAMA (Ciclo Infinito)
# =====================================================================
while True:
    mostrar_menu()
    opcion = input("Elige una opción (1-3): ")

    # Estructura condicional IF, ELIF, ELSE para el control del menú
    if opcion == "1":
        registrar_recepcionista()
    elif opcion == "2":
        mostrar_tabla()
    elif opcion == "3":
        print("\nApagando el simulador de IA... ¡Hasta luego!")
        break # Rompe el ciclo While True y termina el programa
    else:
        print("\n[Error] Opción no válida. Intenta de nuevo con 1, 2 o 3.")
```
